```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
da = pd.read_csv("C:/Users/tannu/Downloads/combined.csv")
da['Frequency']
da.rename( columns ={'Frequency':"Freq"}, inplace= True )
```

```python
X = da["TENG(V)"].values.reshape(-1,1)
  

```

```python
da['Freq'] = da['Freq'].astype(float)
y= da["Freq"].values.reshape(-1,1)
da['Freq'].head()
```

```python

from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


regressor = DecisionTreeRegressor()

regressor.fit(X_train, y_train)
```

```python
#predicting a new value

X = np.array(input("Enter value of Voltage obtained from TENG (V):" ))
X= X.reshape(-1,1)
y_pred = regressor.predict(X)

print("Predicted frequency: % d\n"% y_pred)
```
